Is it possible to predict a songs genre or popularity from data known about it? # CS986 Spotify Classification Problem

This part aims to predict the genre of a song using machine learning techniques.  

The metric used to measure the performance of our final model is accuracy.

## Import and Clean the Data

Firstly, pandas was imported and the training and test sets were read in as dataframes.

```python
#Loading in the training and test data
import pandas as pd

train = pd.read_csv("CS98XClassificationTrain.csv")
test = pd.read_csv("CS98XClassificationTest.csv")
```

We first checked to see if there were any missing values in the data.  The only variable that had missing values was 'top genre'.  Since 'top genre' is the label attribute, and that we will be using supervised learning techniques, all these rows were dropped.  The ID column just represents the row number of each data record and we, therefore, assumed that it had no relation to 'top genre' and it was dropped. 

The frequency of each artist within the dataset was then analysed. Figure 1 shows the vast majority of artists only appeared once in the dataset. Given that 'artist' is categorical data it is very high 'uniqueness' we assumed it would not be useful for analysis. It also would mean that it was quite likely new artists would be in the test data set that we had not seen in the training set.

After being left with just the numeric variables these were scaled and the data was prepared. We applied the same cleaning process to the test data set as the training data set.

```python
#Exploratory analysis
import matplotlib.pyplot as plt
artist_freq = train['artist'].value_counts()
artist_freq.hist()
plt.title(label = 'Fig 1: Histogram of number of songs in dataset per artist');
plt.xlabel('Count of songs in dataset by artist');
plt.ylabel('Freqency');
```

```python
#Cleaning the training data

#Firstly, dropping any rows with incomplete data
incomplete_rows = train[train.isnull().any(axis=1)]
train_dropped = train.drop(incomplete_rows.index)

#Seperating the genre, ID, title and artist columns from the training data
train_labels = train_dropped[["top genre"]]
train_dropped = train_dropped.drop(['top genre','Id', 'title', 'artist'],axis=1) 

#Scaling the variables in train_dropped
from sklearn import preprocessing
std_scaler = preprocessing.StandardScaler()
attribs = list(train_dropped)
train_dropped[attribs] = std_scaler.fit_transform(train_dropped[attribs])

#Train data is now prepared
train_prepared = train_dropped

#Cleaning and scaling the test data for input in the models
test_dropped = test.drop(['Id', 'title', 'artist'], axis=1)
num_attribs = list(train_dropped)

test_dropped[num_attribs] = std_scaler.fit_transform(test_dropped[num_attribs])
test_prepared = test_dropped
```

```python
#Histogram of genre frequency
top_genre_freq = train['top genre'].value_counts()
top_genre_freq.plot.bar(figsize = (11,4))
plt.title(label = 'Fig 2: Frequency of top genre in training set');
plt.ylabel('Frequency');plt.xlabel('top genre');
```

## Modelling

Choosing a classifier for this problem is quite hard. The number of data points is not very big, 438 after cleaning. Also, figure 2 shows that many of the 81 genres have only very few entries in the data. We also know that the genre a song is classified as can be a matter of opinion and different genres may overlap significantly. 

The fist classifiers used was a logistic regressor enveloped in a one versus rest classifier. Logistic regression was used as it is a simple and reliable model for classification and would be a good first model to try. There are 81 unique values in 'top genre' so we need a multiclass classifier. To apply logistic regression as a multiclass classifier it was used within a one verses rest (OVR) classifier. 

The second model used was a linear support vector machine (SVM) classifier.  This was used as the data we are dealing has a relatively small size and it is fairly complex.  Also, in tidying the data it was scaled so it can be used with this classifier.  Although SVM works well with small and less complicated datasets like the one we are working with, we suspect that the genre classes may be somewhat overlapping and are therefore expecting mixed results from linear SVM.

The third model used was a voting classifier model that made use of the first 2 models.  This was done as we expected that by combining the 2 models we already have would result in better classifications overall.

```python
#Loading in and naming the classifier models we will use

from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import VotingClassifier

clf = LogisticRegression()
ovr_clf = OneVsRestClassifier(LogisticRegression())
svm_clf = LinearSVC()
voting_clf = VotingClassifier(estimators=[('OVR', ovr_clf), ('SVM', svm_clf)],
                              voting='hard')
```

```python
#suppress convergence warning
import warnings
with warnings.catch_warnings():
  warnings.simplefilter("ignore")

  ##Fitting the models using the training data
  #OVR model
  ovr_clf.fit(train_prepared, train_labels)
  ovr_predictions = ovr_clf.predict(test_prepared)
  #SVM model
  svm_clf.fit(train_prepared, train_labels)
  svm_predictions = svm_clf.predict(test_prepared)
  #Voting model
  voting_clf.fit(train_prepared, train_labels)
  voting_predictions = voting_clf.predict(test_prepared)
```

For each of the 3 models the accuracy score was predicted.  This is predicted as the number of correct predictions divided by the total number of predictions.  The OVR model had a accuracy score of 0.39954, the SMV model had 0.43607 and the voting classifier had 0.40867.  It is expected that the accuracy score for the SVM would be better than that for the OVR because the OVR is a simple model.  It also makes sense that the voting classifier has an accuracy score in the middle because it contains these 2 models.

```python
#Calculating the accuracy scores for the models using the training set

from sklearn.metrics import accuracy_score

ovr_predictions_train = ovr_clf.predict(train_prepared)
print('OVR', accuracy_score(train_labels, ovr_predictions_train))

svm_predictions_train = svm_clf.predict(train_prepared)
print('SVM', accuracy_score(train_labels, svm_predictions_train))

voting_predictions_train = voting_clf.predict(train_prepared)
print('Voting', accuracy_score(train_labels, voting_predictions_train))
```

Finally, the predictions for each of the models were made into csv files so that we could get an accuracy score for each of the models when using the test data when the predictions were uploaded to Kaggle.  The scores for the OVR and SVM models were both 0.28571 and the score for the voting classifier was 0.30357.  These values are lower than the accuracy scores when using the training set.  This is to be expected as the models were fit using the training set so the accuracy scores would be higher.

```python
#Making the csv files containing the final predictions for the voting model

Voting_submission = pd.DataFrame({'Id':test['Id'],
                                  'top genre':voting_predictions})
filename = 'Classification Predictions Voting.csv'
Voting_submission.to_csv(filename,index=False)
```

##Conclusion

We had also tried some other solutions. Decision trees and random forest classifiers were used but produced very poor results on the final test data. We believe that random forest and decision trees performed poorly for a few reasons. Firstly, the label, 'top genre', are likely to have been overlapping somewhat, which these models do not deal with well. The models were also susceptible to overfitting due to the small dataset size and the relatively large number of labels, meaning that each 'leaf' would have very few instances on it, reducing the model's performance. 

We had also tried fitting all our models to another dataset where instances containing unique (count of 1 in dataset) artists were completely removed or replaced by the placeholder 'other'. This was done because we expected that correctly predicting these small frequency genres in the test set is very unlikely, and that including them may produce biases. The accuracy of the test results was either the same as with the full dataset or worse, so the full dataset was used in the final models.

In conclusion, our final predictions used will be the voting classifier predictions as these returned the highest accuracy score when submitted to the online competition achieving a value of 0.30357.  We believe that this accuracy score would be higher if the frequency of genres in the training set were more evenly spread.  As it is the genres adult standards, album rock and dance pop all have many instances as shown in figure 2, whereas the majority of genres have very few instances.  This causes the models to be more inclined to predicting the more frequently appearing genres as they have more data about them.  This also meant that it was hard to find a balance between creating a model that was general enough to make accurate predictions on the test data, whilst also being careful to not create a model that was too specific and overfitting on the training data causing it to be poor at making predictions from the test data.

# CS986 Spotify Regression Problem 

This part of the notebook aims to build a machine learning model to predict an integer popularity value of a song.

The metric used to assess the performance of the model is the root mean square error (RMSE).

## Importing and Cleaning Data

```python
#Loading in the training and test data
train_reg = pd.read_csv("CS98XRegressionTrain.csv")
test_reg = pd.read_csv("CS98XRegressionTest.csv")
```

'ID' was removed as this is an index and artists was not used due to the number of unique entries which are unlikely to generalise to test datasets (see figure 1).

```python
train_reg = train_reg.rename(columns={'pop': 'popularity'})

features = train_reg.loc[:,"top genre":"spch"]
label = train_reg[["popularity"]]
```

The 'top genre' column contains categorical data which is not well-handled by regression algorithms, so this column was one-hot encoded.

```python
cat_features = features.loc[:,"top genre"]
cat_dummies = pd.get_dummies(cat_features, drop_first=True)

features = features.drop(["top genre"], axis=1)
```

To better suit a variety of models, the data was scaled using the Standardscaler function.

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
features = pd.DataFrame(std_scaler.fit_transform(features), columns = features.columns)
features.head()

features = pd.concat([label, features, cat_dummies], axis=1)
```

15 missing values were identified in the 'top genre' column. 

```python
train_reg.isna().sum()
```

Missing values were handled using the KNNImputer function which utilises the mean value from n_neighbors nearest neighbors found in the training set. Samples are close if they have similar features.


```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
features = pd.DataFrame(imputer.fit_transform(features),columns = features.columns)
```

```python
X = features.loc[:, "year":"yodeling"]
y = features["popularity"]
```

## Exploratory Data Analysis

The features in this dataset are not highly correlated with the popularity, only 7 of the 95 features have an absolute correlation greater than 0.2, which may make it difficult to build an accurate model.

```python
corr_matrix = features.corr()
corr_matrix[abs(corr_matrix["popularity"])>0.2]["popularity"]
```

## Model

Gradient boosting regression was chosen as it incorporates decision trees which are a powerful algorithm for handling complex data. Gradient boosting regression fits a decision tree to the residuals of the previous model which can help reduce overfitting. We also tried support vector regression due to its suitability for complex small to medium sized datasets. Grid searches were used on these two models and on a selection of other models that would be used for stacking. These include ridge regression, extreme gradient boosting, k-nearest neighbours algorithm for regression, and random forest regression.

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
GBR = GradientBoostingRegressor()
parameters = {'n_estimators' : [100,250,500,750,1000],
              'max_depth'    : [5,6,7,8,9,10]
              }
grid_GBR = GridSearchCV(estimator=GBR, param_grid = parameters, cv = 2, n_jobs=-1)
grid_GBR.fit(X, y)

print(" Results from Grid Search " )
print("\n The best estimator across ALL searched params:\n",
      grid_GBR.best_estimator_)
print("\n The best score across ALL searched params:\n",
      grid_GBR.best_score_)
print("\n The best parameters across ALL searched params:\n",
      grid_GBR.best_params_)
```

An example of parameter tuning for gradient boosting regression using the gridsearch function. The best parameters found were max_depth = 5 and n_estimators = 100. The number of estimators is the number of concurrent decision trees in the model and the maximum depth is the maximum number of nodes for each tree. After parameter tuning had been applied to each model they were feed into a stacking function.

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Table(header=dict(values=['Model', 'Mean Absolute Error']),
                 cells=dict(values=[["Ridge Regression", "Support Vector Regression", "Gradient Boosting Regression",
                                     "K-Nearest Neighbours Regression", "Random Forest Regression",
                                     "Extreme Gradient Boosting Regression"], [8.58252928, 8.85045148, 8.35194036,
                                                                               8.69103039, 9.32457628, 8.92493949]]),
                               columnwidth = 2
                               )])
fig.update_layout(title_text = "Stacking Output", title_x=0.5, )
#fig.show()
```

Gradient boosting regression has the lowest mean absolute error and therefore is the best performing model.

```python
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m],
                                               y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
        
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   
    plt.xlabel("Training set size", fontsize=14) 
    plt.ylabel("RMSE", fontsize=14)  
    plt.title("Fig 4")            

plot_learning_curves(grid_GBR.best_estimator_, X, y)






##space added for figure to print on page##
```



A learning curve, plotted using pyplot and 5-fold cross-validation, found that the model was still overfitting so we attempted to re-fit the model using fewer features and removing outliers with z-scores greater than a variety of thresholds but this had little effect on the amount of overfitting. Genres were also combined as a method of reducing the number of features. For example, all "top genre" entries containing 'pop' were grouped together, this also resulted in minimal improvements.

```python
final_model =  GradientBoostingRegressor(max_depth = 5, 
                                         n_estimators = 100,
                                         random_state = 42)

final_model.fit(X, y)

final_model_predictions = final_model.predict(X)

final_model_mse = mean_squared_error(y, final_model_predictions)
final_model_rmse = np.sqrt(final_model_mse)
final_model_rmse
```

The final model had an RMSE of 3.304 on the training dataset.

## Conclusion

The chosen gradient boosting regressor had an RMSE of 7.89556 on the test dataset. The performance is likely worse than on training data because of the model not generalising well due to finding patterns in the training data that are not present in the test data. Furthermore, estimates of popularity tended to cluster around the mean of the training set, thereby providing a particularly poor performance on outliers in the test set.

Extensive parameter tuning was performed on a wide variety of models and yielded little improvement in the regressor's performance. Further efforts may therefore be best directed towards feature engineering, potentially finding more suitable transformations for the initial data, or possibly combining these to produce new features. Some of the problems with overfitting may be dealt with by training the model on a larger subset of the original playlists.
